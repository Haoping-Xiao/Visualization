```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as md

data=pd.read_csv('./BTCvsSP500.csv')

data=data.loc[:,['date','BTC','SP500']]
data['date'] = pd.to_datetime(data['date'], format = '%Y-%m-%d')
data.set_index('date',inplace=True)
data.iloc[:-8,:].plot.line()
plt.xticks(rotation=0)
plt.suptitle('price in the past few years',fontsize=15)
plt.xlabel('date')
plt.ylabel('price')
```

```python
# plt.figure(figsize=(20,6))
figure, axes = plt.subplots(1, 2)
data.plot.line(ax=axes[0],figsize=(18,6),title='price in the past few years')
plt.xticks(rotation=0)
# plt.title('price in the past few years',fontsize=15)
plt.xlabel('date')
plt.ylabel('price')
data.iloc[-183:,:].plot.line(ax=axes[1],figsize=(18,6),title='price in the past few months')
# plt.title('price in the past few months',fontsize=15)
plt.xlabel('date')
plt.ylabel('price')
plt.xticks(rotation=45)
```

```python
data.iloc[-183:,:].plot.line()
plt.suptitle('price in the past few months',fontsize=15)
plt.xlabel('date')
plt.ylabel('price')
plt.xticks(rotation=45)
```

```python
fig=data.iloc[-183:,:].hist(bins=10)
start_date=data.iloc[-183,:].name
end_date=data.iloc[-1,:].name
start_date=start_date.strftime('%Y-%m-%d')
end_date=end_date.strftime('%Y-%m-%d')
plt.suptitle('price distribution in the past few months \n from {} to {}'.format(start_date,end_date)
             ,fontsize=15,y=1.1)
fig[0][0].set_xlabel('price')
fig[0][1].set_xlabel('price')
```

```python
start_date=data.iloc[0,:].name
end_date=data.iloc[-1,:].name
fig=data.hist(bins=10)
plt.suptitle('price distribution in the past few years \n from {} to {}'.format(start_date,end_date)
             ,fontsize=15,y=1.1)
fig[0][0].set_xlabel('price')
fig[0][1].set_xlabel('price')
```
