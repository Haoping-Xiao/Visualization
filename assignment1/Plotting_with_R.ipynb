```r
library(ggplot2)
labels <-c('Under $100', 'About $100', '$100 TO $200', '$200 TO $300' , 'MORE THAN $300')
sizes <- c(22, 26, 39, 10, 3)
props <- c('22%', '26%', '39%', '10%', '3%')
data<-data.frame(Category =labels,
                 value=sizes)



piechart<-ggplot(data,aes(x="",y=value,fill=Category))+
          ggtitle("How much do you spend on groceries every week?")+
          geom_bar(stat="identity", width=1) +
          coord_polar("y", start=0) +
          theme_void() +# remove background, grid, numeric labels 
          theme(plot.title = element_text(lineheight=1, face="bold", hjust = 0.5, size=15))+
          geom_text(aes(label = props),position = position_stack(vjust = 0.5),color = "white", size=6) 
piechart
```
